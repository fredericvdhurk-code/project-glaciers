```python
import pandas as pd
import statsmodels.api as sm
```

```python
mass_balance_df = pd.read_csv('project-glaciers/data/mass_balance_hy.csv')
weather_davos_hy = pd.read_csv('project-glaciers/data/weather_data_davos_hy.csv')
weather_davos_seasonal = pd.read_csv('project-glaciers/data/weather_data_davos_seasonal.csv')
silvretta_mb = mass_balance_df[mass_balance_df['glacier name'] == 'Silvrettagletscher']
silvretta_mb = silvretta_mb[['annual mass balance (mm w.e.)']]
silvretta_mb_w_s = mass_balance_df[mass_balance_df['glacier name'] == 'Silvrettagletscher']
silvretta_mb_w = silvretta_mb_w_s[['winter mass balance (mm w.e.)']]
silvretta_mb_s = silvretta_mb_w_s[['summer mass balance (mm w.e.)']]
# Even rows → winter (0, 2, 4, ...)
weather_davos_winter = weather_davos_seasonal.iloc[::2, :]  

# Odd rows → summer (1, 3, 5, ...)
weather_davos_summer = weather_davos_seasonal.iloc[1::2, :]

weather_davos_summer = weather_davos_summer.reset_index(drop=True)
weather_davos_winter = weather_davos_winter.reset_index(drop=True)
weather_davos_hy
```

```python
# Merge on year
silvretta_reg_t = silvretta_mb.merge(weather_davos_hy, left_index=True, right_index=True)

# Define predictors and target
X = silvretta_reg_t[['mean temperature (°C)', 
                       'mean max daily temperature (°C)',
                       'mean min daily temperature (°C)'
                  ]]
y = silvretta_reg_t['annual mass balance (mm w.e.)']

# Add intercept for statsmodels
X = sm.add_constant(X)

# Fit regression
model = sm.OLS(y, X).fit()

print(model.summary())
```
