```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

mass_balance_hy_df = pd.read_csv('project-glaciers/data/mass_balance_hy.csv')
```

```python
import seaborn as sns
print(sns.__version__)
```

```python
# Rhonegletscher mass balance visualization


rhone_mb_hy_df = mass_balance_hy_df[mass_balance_hy_df['glacier name'] == 'Rhonegletscher']
rhone_mb_hy_df = rhone_mb_hy_df[['start date of observation (yyyy-mm-dd)',
                                     'end date of observation (yyyy-mm-dd)',
                                     'annual mass balance (mm w.e.)'
                                ]]

rhone_mb_hy_df['date'] = pd.to_datetime(rhone_mb_hy_df['end date of observation (yyyy-mm-dd)'])
rhone_mb_hy_df = rhone_mb_hy_df[['date',
                                     'annual mass balance (mm w.e.)'
                                ]]
rhone_mb_hy_df = rhone_mb_hy_df.set_index('date')
rhone_mb_hy_df = rhone_mb_hy_df.rename(columns={'annual mass balance (mm w.e.)': 'rhone mass balance'})


rhone_mb_hy_df
```

```python
aletsch_mb_hy_df = mass_balance_hy_df[mass_balance_hy_df['glacier name'] == 'Grosser Aletschgletscher']
aletsch_mb_hy_df = aletsch_mb_hy_df[['start date of observation (yyyy-mm-dd)',
                                         'end date of observation (yyyy-mm-dd)',
                                         'annual mass balance (mm w.e.)'
                                    ]]

aletsch_mb_hy_df['date'] = pd.to_datetime(aletsch_mb_hy_df['end date of observation (yyyy-mm-dd)'])
aletsch_mb_hy_df = aletsch_mb_hy_df[['date',
                                         'annual mass balance (mm w.e.)'
                                    ]]
aletsch_mb_hy_df = aletsch_mb_hy_df.set_index('date')
aletsch_mb_hy_df = aletsch_mb_hy_df.rename(columns={'annual mass balance (mm w.e.)': 'aletsch mass balance'})

aletsch_mb_hy_df
```

```python
# Merge the two DataFrames on their indices (years)
combined_df = aletsch_mb_hy_df.join(rhone_mb_hy_df, how = 'inner')  # 'inner' keeps only common years
print("\nCombined Data (Common Years):")
print(combined_df)
```

```python

# Scatter plot
plt.figure(figsize = (10, 6))
sns.scatterplot(
    data = combined_df,
    x = 'aletsch mass balance',
    y = 'rhone mass balance',
    hue = combined_df.index.year,  # Color points by year
    palette = 'viridis',
    s = 75
)

plt.title('Rhone vs. Aletsch Glacier Mass Balance (Common Years)')
plt.xlabel('Aletsch Glacier Mass Balance')
plt.ylabel('Rhone Glacier Mass Balance')
plt.grid(True)
plt.show()
```

```python
# Extract x and y
x = combined_df[['aletsch mass balance']].values
y = combined_df['rhone mass balance'].values

# Fit the model
model = LinearRegression()
model.fit(x, y)

# Predictions
y_pred = model.predict(x)

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=combined_df,
    x = 'aletsch mass balance',
    y = 'rhone mass balance',
    hue = combined_df.index.year,
    palette = 'viridis',
    s = 75
)
plt.plot(x, y_pred, color = 'red', linewidth = 1, label = 'Linear Regression')
plt.title('Rhone and Aletsch Glacier Mass Balance (Common Years)')
plt.xlabel('Aletsch Glacier Mass Balance')
plt.ylabel('Rhone Glacier Mass Balance')
plt.grid(True)
plt.legend()
plt.show()
print(model.coef_[0])  
print(model.intercept_)
```

```python
# Chessjengletscher mass balance visualization


chessjengletscher_mb_hy_df = mass_balance_hy_df[
    mass_balance_hy_df['glacier name'].isin(
        ['Chessjengletscher', 'Chessjengletscher NW']
    )
]
chessjengletscher_mb_hy_df = chessjengletscher_mb_hy_df[['start date of observation (yyyy-mm-dd)',
                                                             'end date of observation (yyyy-mm-dd)',
                                                             'annual mass balance (mm w.e.)'
                                                        ]]
chessjengletscher_mb_hy_df = chessjengletscher_mb_hy_df.reset_index(drop=True)


#aletsch_mb_hy_df_2007 = aletsch_mb_hy_df.drop(aletsch_mb_hy_df.index[0:93])


# Convert date columns to datetime
chessjengletscher_mb_hy_df['date'] = pd.to_datetime(chessjengletscher_mb_hy_df['end date of observation (yyyy-mm-dd)'])
chessjengletscher_mb_hy_df = chessjengletscher_mb_hy_df[['annual mass balance (mm w.e.)', 'date']]

chessjengletscher_mb_hy_df = chessjengletscher_mb_hy_df.set_index('date')
chessjengletscher_mb_hy_df = chessjengletscher_mb_hy_df.rename(columns={'annual mass balance (mm w.e.)': 'chessjengletscher mass balance'})
```

```python
hohlaubgletscher_mb_hy_df = mass_balance_hy_df[mass_balance_hy_df['glacier name'] == 'Hohlaubgletscher']
hohlaubgletscher_mb_hy_df = hohlaubgletscher_mb_hy_df[['start date of observation (yyyy-mm-dd)',
                                                           'end date of observation (yyyy-mm-dd)',
                                                           'annual mass balance (mm w.e.)'
                                                      ]]
hohlaubgletscher_mb_hy_df = hohlaubgletscher_mb_hy_df.reset_index(drop=True)


#aletsch_mb_hy_df_2007 = aletsch_mb_hy_df.drop(aletsch_mb_hy_df.index[0:93])


# Convert date columns to datetime
hohlaubgletscher_mb_hy_df['date'] = pd.to_datetime(hohlaubgletscher_mb_hy_df['end date of observation (yyyy-mm-dd)'])
hohlaubgletscher_mb_hy_df = hohlaubgletscher_mb_hy_df[['annual mass balance (mm w.e.)', 'date']]
hohlaubgletscher_mb_hy_df = hohlaubgletscher_mb_hy_df.set_index('date')
hohlaubgletscher_mb_hy_df = hohlaubgletscher_mb_hy_df.rename(columns={'annual mass balance (mm w.e.)': 'hohlaubgletscher mass balance'})
```

```python
# Merge the two DataFrames on their indices (years)
combined_df_2 = hohlaubgletscher_mb_hy_df.join(chessjengletscher_mb_hy_df, how = 'inner')  # 'inner' keeps only common years
print("\nCombined Data (Common Years):")
print(combined_df_2)
```

```python
# Extract x and y
x = combined_df_2[['hohlaubgletscher mass balance']].values
y = combined_df_2['chessjengletscher mass balance'].values

# Fit the model
model = LinearRegression()
model.fit(x, y)

# Predictions
y_pred = model.predict(x)

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=combined_df_2,
    x = 'hohlaubgletscher mass balance',
    y = 'chessjengletscher mass balance',
    hue = combined_df_2.index.year,
    palette = 'viridis',
    s = 75
)
plt.plot(x, y_pred, color = 'red', linewidth = 1, label = 'Linear Regression')
plt.title('Chessjengletscher and Hohlaubgletscher Glacier Mass Balance (Common Years)')
plt.xlabel('Hohlaubgletscher Glacier Mass Balance')
plt.ylabel('Chessjengletscher Glacier Mass Balance')
plt.grid(True)
plt.legend()
plt.show()
print(model.coef_[0])  
print(model.intercept_)
```
