```python
import pandas as pd
import os
import numpy as np
```

```python
davos = pd.read_csv('project-glaciers/data/weather_data_davos_monthly.csv')
sion = pd.read_csv('project-glaciers/data/weather_data_sion_monthly.csv')
sion_dev_t = sion[['th9120mv']]
davos_dev_t = davos[['th9120mv']]
folder_path = "project-glaciers/data"
```

```python
## Sion
sion_dev_t = sion_dev_t.reset_index(drop=True)

# Step 1: Add a helper column to group every 12 rows
sion_dev_t['year'] = sion_dev_t.index // 12

# Step 2: Add a helper column for the position within each group
sion_dev_t['month'] = sion_dev_t.index % 12

# Step 3: Pivot the DataFrame
sion_monthly_temp_dev = sion_dev_t.pivot(index = 'year',
                                   columns = 'month',
                                   values = 'th9120mv')


# Define the new column names
monthly_observations = [
    'october_td', 'november_td', 'december_td',
    'january_td', 'february_td', 'march_td',
    'april_td', 'may_td', 'june_td',
    'july_td', 'august_td', 'september_td'
]

# Rename the columns
sion_monthly_temp_dev.columns = monthly_observations

# Create the new column with the start date of the hydrological year
sion_monthly_temp_dev['hydrological year'] = [
    f"{1914 + i}-{1915 + i}" for i in range(len(sion_monthly_temp_dev))
]

sion_monthly_temp_dev = sion_monthly_temp_dev.reset_index(drop=True)

sion_monthly_temp_dev


## Davos

davos_dev_t = davos_dev_t.reset_index(drop=True)

# Step 1: Add a helper column to group every 12 rows
davos_dev_t['year'] = davos_dev_t.index // 12

# Step 2: Add a helper column for the position within each group
davos_dev_t['month'] = davos_dev_t.index % 12

# Step 3: Pivot the DataFrame
davos_monthly_temp_dev = davos_dev_t.pivot(index = 'year',
                                   columns = 'month',
                                   values = 'th9120mv')



# Rename the columns
davos_monthly_temp_dev.columns = monthly_observations

# Create the new column with the start date of the hydrological year
davos_monthly_temp_dev['hydrological year'] = [
    f"{1914 + i}-{1915 + i}" for i in range(len(davos_monthly_temp_dev))
]

davos_monthly_temp_dev = davos_monthly_temp_dev.reset_index(drop=True)

davos_monthly_temp_dev
```

```python
## SION

sion_precipitation = sion[['rhs150m0', 'rh9120mv']].copy()
sion_precipitation['rh9120mv'] = sion_precipitation['rh9120mv'] / 100
sion_precipitation['precipitation_dev'] = sion_precipitation['rhs150m0'] - (sion_precipitation['rhs150m0'] / sion_precipitation['rh9120mv'])
sion_precipitation['precipitation_dev'] = sion_precipitation['precipitation_dev'].round(1)
sion_precipitation_dev = sion_precipitation[['precipitation_dev']].copy()

# Step 1: Add a helper column to group every 12 rows
sion_precipitation_dev['year'] = sion_precipitation_dev.index // 12

# Step 2: Add a helper column for the position within each group
sion_precipitation_dev['month'] = sion_precipitation_dev.index % 12

# Step 3: Pivot the DataFrame
sion_precipitation_dev = sion_precipitation_dev.pivot(index = 'year',
                                   columns = 'month',
                                   values = 'precipitation_dev')

# Define the new column names
monthly_observations_pd = [
    'october_pd', 'november_pd', 'december_pd',
    'january_pd', 'february_pd', 'march_pd',
    'april_pd', 'may_pd', 'june_pd',
    'july_pd', 'august_pd', 'september_pd'
]
# Rename the columns
sion_precipitation_dev.columns = monthly_observations_pd

# Create the new column with the start date of the hydrological year
sion_precipitation_dev['hydrological year'] = [
    f"{1914 + i}-{1915 + i}" for i in range(len(sion_precipitation_dev))
]

sion_precipitation_dev = sion_precipitation_dev.reset_index(drop=True)




## DAVOS

davos_precipitation = davos[['rhs150m0', 'rh9120mv']].copy()
davos_precipitation['rh9120mv'] = davos_precipitation['rh9120mv'] / 100
davos_precipitation['precipitation_dev'] = davos_precipitation['rhs150m0'] - (davos_precipitation['rhs150m0'] / davos_precipitation['rh9120mv'])
davos_precipitation['precipitation_dev'] = davos_precipitation['precipitation_dev'].round(1)
davos_precipitation_dev = davos_precipitation[['precipitation_dev']].copy()

# Step 1: Add a helper column to group every 12 rows
davos_precipitation_dev['year'] = davos_precipitation_dev.index // 12

# Step 2: Add a helper column for the position within each group
davos_precipitation_dev['month'] = davos_precipitation_dev.index % 12

# Step 3: Pivot the DataFrame
davos_precipitation_dev = davos_precipitation_dev.pivot(index = 'year',
                                   columns = 'month',
                                   values = 'precipitation_dev')

# Rename the columns
davos_precipitation_dev.columns = monthly_observations_pd

# Create the new column with the start date of the hydrological year
davos_precipitation_dev['hydrological year'] = [
    f"{1914 + i}-{1915 + i}" for i in range(len(davos_precipitation_dev))
]

davos_precipitation_dev = davos_precipitation_dev.reset_index(drop=True)

davos_precipitation_dev
```

```python
sion_monthly_temp_dev.to_csv(os.path.join(folder_path, "weather_dev_sion_temp.csv"), index = False)
davos_monthly_temp_dev.to_csv(os.path.join(folder_path, "weather_dev_davos_temp.csv"), index = False)
sion_precipitation_dev.to_csv(os.path.join(folder_path, "weather_dev_sion_prec.csv"), index = False)
davos_precipitation_dev.to_csv(os.path.join(folder_path, "weather_dev_davos_prec.csv"), index = False)
```
